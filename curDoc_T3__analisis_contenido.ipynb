```python
import openai
from docx import Document
import pandas as pd
import os
from openai import OpenAI
from google.colab import userdata
KEYX= userdata.get('OAIK_JCMV')
client = OpenAI(api_key=KEYX)
```

```python
ruta = 'drive/My Drive/00 2023_all/2023 Cursos/Capacidades Cursos/cursos_LLM_codigos/'
print(os.listdir(ruta))
```

```python
archivo = 'parrafos_quijote.xlsx'
```

```python
base_textos = pd.read_excel(ruta + archivo)
```

```python
base_textos
```

```python
prompt_sistema = "Eres una experta en identificación de patrones linguisticos en textos literarios."

terminos = """
Los términos o definiciones a tomar en cuenta para el análisis son:
Actividades Mentales: Cuando se alude a algun ejercicio mental de los protagonistas del texto, como razonamientos, pensamientos, sueños, ideas u otros
Actividades Físicas: Cuando se alude a algún movimiento o ejercicio de fuerza física de los protagonistas
"""

pedido = """
Tomando en cuenta el texto a analizar, detectar y reportar dos listas: en una lista separada por comas cuáles son las actividades mentales físicas detectadas, y en otra lista separada por comas cuáles son las actividades mentales detectadas
"""
```

```python
def extractos_linguisticos(texto):

    prompt_usuario = prompt_sistema + terminos + pedido + f"""El texto a analizar es: \"\"\"{texto}\"\"\" """

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {'role': 'system', 'content': prompt_sistema},
            {"role": "user", "content": prompt_usuario},
            ],
        temperature=0.7,
        max_tokens=350,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0
      )

    return response.choices[0].message.content
```

```python
respuestas = []
for j, row in base_textos.iterrows():
    print(row.item, end=' ')
    tx_x = row.fragmento
    respuestas.append(extractos_linguisticos(tx_x))
    print('OK')

resp_df = pd.DataFrame(respuestas)
```
